# Symmetric Difference

```python
a={1,2,3,4,5}
b={4,5,6,7,8}
c={8,9,10}

print(a)
print(b)
print(c)
```

```python
a.symmetric_difference(b)    # Will display uncommon items
```

```python
a ^ c      # Symmetric difference is indicated by ^ symbol
```

# Symmetric Difference Update

```python
a={1,2,3,4,5}
b={4,5,6,7,8}
c={8,9,10}

print(a)
print(b)
print(c)
```

```python
a.symmetric_difference_update(b)   # it is removing common values in 'a' & 'b' and updating the 'a' set into new values 
print(a) 
```

```python
b.symmetric_difference_update(c)
print(b)
```

```python
print(a)    # you can see both a and b values in set are changed
print(b)
print(c)
```

# Superset, Subset, Disjoint

```python
# Superset is like a dad 
# Subset is like a son   # all values in subset must be present in superset then only it is dad and son relation
# Disjoint is like a Neighbour   # no values must be common between disjoint and superset then it is a neighbour
```

```python
s4={1,2,3,4,5,6,7,8,9}
s5={4,5,6,7,8}
s6={10,20,30,40}

print(s4)
print(s5)
print(s6)      # all values of s5 must be present in s4 thats why s4 is superset of s5
```

```python
s4.issuperset(s5)   # s4 is the dad of s5 = true
```

```python
s5.issubset(s4)    # s5 is the son of s4 = true
```

```python
s6.isdisjoint(s4)   # s6 is the neighbour of s4 = true
```

```python
s6.issuperset(s4)   # s6 is the dad of s4 = false
```

```python
s6.issubset(s4)     # s6 is the son of s4 = false 
```

```python
s7={1,2,3,4,5,6,7,8,9}
s8={15,25,35,9}
s9={10,20,30,40}

print(s7)
print(s8)     # in s8 set only value 9 is common between s7 and s8, but to be a superset all values must be coomon
print(s9)
```

```python
s7.issuperset(s8)    # only value 9 is present that why it is false
```

```python
s8.issubset(s7)     
```

```python
s9.isdisjoint(s8)   # no value should match then it is a neighbour
```
