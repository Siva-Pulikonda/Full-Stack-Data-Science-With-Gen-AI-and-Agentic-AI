# Swap Variable

```python
a=5
b=6

print(a)
print(b)
```

```python
a=b     # Here instead of Swaping we lost a variable
b=a 

print(a)
print(b)
```

```python
a1=7
b1=8

print(a1)
print(b1)
```

```python
temp=a1    # by using temp as a third variable we are swaping
a1=b1
b1=temp

print(a1)
print(b1)
```

```python
# without using third variable swap these two 

a2=9
b2=10

print(a2)
print(b2)
```

```python
a2=a2+b2   # Swapping Without 3rd variable
b2=a2-b2
a2=a2-b2

print(a2)
print(b2)
```

```python
# easiest way to swap is 
a3=11
b3=12

print(a3)
print(b3)
```

```python
a3,b3=b3,a3

print(a3)
print(b3)
```

# Optimization

```python
import math as m
sqrt_result = m.sqrt(225)    # here we are writing more lines of code, to reduce that we are using optimization
pow_result = m.pow(5, 3)

print(sqrt_result)
print(pow_result)
```

```python
from math import sqrt,pow    # we have reduced the no of lines of code

print(pow(3,3))
print(sqrt(169))
```
