# 1) Array Creation Function

```python
import numpy as np
```

```python
# Create an array from list   
a=np.array([1,2,3])              # we convert list argument to array
print("Array a:", a)
```

```python
# Create an array with evenly spaced values
b=np.arange(0,10,2)  # Values from 0 to 10 with step 2 
print("Array b:", b)
```

```python
# Create an array with odd values
c=np.arange(0,10,3)
print("Array c:",c)
```

```python
# create an array filled with zeros
d=np.zeros((2,3))  # 2*3 Array of zeros
print("Array d:\n",d)
```

```python
# create an array filled with zeros
d=np.zeros((2,3),dtype=int)  # 2*3 Array of zeros
print("Array d:\n",d)
```

```python
# create an array filled with ones with 3*2 matrix
e=np.ones((3,2),dtype=int)
print("Array e:\n",e)
```

```python
#Create a identity 4*4 matrix
f=np.eye(4)
print("identity matrix f:\n",f)
```

# 2) Array Manipulation function

```python
# Reshape an Array Reshape into 1*3 ( 1 row three columns)
a1=np.array([1,2,3])
reshaped=np.reshape(a1,(1,3))
print("Reshaped array:",reshaped)
```

```python
# Reshape an Array Reshape into 3*1( 3 row,1 columns)
a1=np.array([1,2,3])
reshaped=np.reshape(a1,(3,1))
print("Reshaped array:",reshaped)
```

```python
#Flatten a 1D array
f1=np.array([[1,2],[3,4]])
flattened= np.ravel(f1)
print("Flattened Array;",flattened)
```

```python
# Transpose an array
e1=np.array([[1,2],[3,4]])
transposed= np.transpose(e1)
print("Transposed Array:\n", transposed)
```

```python
# Stacks Array vertically
a2=np.array([1,2])
b2=np.array([3,4])
c2=np.array([5,6])
stacked=np.vstack([a2,b2,c2])
print("stacked arrays:\n",stacked)
```
