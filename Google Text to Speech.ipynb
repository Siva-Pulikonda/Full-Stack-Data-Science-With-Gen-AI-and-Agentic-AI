```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''In a small village, a curious little rabbit named Chintu loved exploring new places. 
One day, he found a shiny golden carrot deep in the forest that could talk and grant wishes. Excited, Chintu wished for endless carrots, 
but soon the forest was filled with so many carrots that no one could move. Realizing his mistake, he wished everything back to normal and 
learned an important lesson. From then on, Chintu used his wishes wisely and always thought before asking for more..''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')  # if wav is not working then change extension.mp3
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= True)
```
