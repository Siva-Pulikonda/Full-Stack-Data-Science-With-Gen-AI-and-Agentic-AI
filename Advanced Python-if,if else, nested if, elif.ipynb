```python
#  Core Rule of if-else

#  Python does NOT care about what you print
#  It only checks the condition (True or False)
#  if True  → run IF block
#  if False → run ELSE block
```

```python
if True:
    print('Data Science')
else:
 print('Agentic AI')     # Here Condition is true so python runs if block
```

```python
if False:
    print('Data Science')
else:
 print('Agentic AI')     # Here Condition is False so python skips if and runs else 
```

```python
if False:                     # if condition is false so python doesnot execute print('Agentic AI')
    print('Agentic AI')       # it goes to else and print('Data Science') 
else:
 print('Data Science')  


# if you interchange the values you think the answer is Agenti AI    
# Here you are thinking like false means first print should run that is incorrect 
# False Means it should skip if block and print else block so the output os Data Science
```

# Print Even & Odd Numbers

```python
x = 4
r = x % 2

if r==0:
    print('Even Number')
```

```python
x = 5
r = x % 2

if r==1:
    print('Odd Number')
```

```python
# Nested if

x = 11
r = x % 2

if r==0:
    print('Even Number')
if r==1:
    print('Odd Number')     # Multiple If s not Recommended so we use if Else
```

```python
# if else is better than nested if

x = 10 
r = x % 2

if r==0:
    print('Even Number')
else:
    print('Odd Number') 
```

```python
# Two conditions 
# 1) check number is even or odd
# 2) check number is greater or smaller
```

```python
x =7
r = x % 2

if r == 0:
    print('Even Number')        # IF, if condition pass this line 
    
    if x>5:
        print('Greather Number')   # then it will go to this line 
    else:
        print('Lesser Number')
        
else:
    print('Odd Number')         # if condition fails directly comes to this line
```
