# While Loop

```python
i=1          # initializing  # incremental print data science 5 times 
while i<=5:  # condition
    print('Data Science')
    i=i+1    # incremantal
```

```python
i=5              # Decremental print data science 5 times 
while i>=1:
    print('Data Science')
    i=i-1
```

```python
i=1         
while i<=5:  
    print('Data Science:',i)  # i will give the no of times the result is printing
    i=i+1    

# using this code anyone can understand you used incremental or decremental
```

# Nested While Loop

```python
# Nested While Loop

i=1 
while i<=5:
    print('Data Science')     # Outer Loop is Condition 1
    j=1
    while j<=4:
        print('Technology')   # inner loop is condition 2
        j=j+1                 # always remember first inner loop will run then outerloop will run
    i=i+1
    print()
```

```python
# if condition 1 satisfies it will go to condition 2, condition 2 will run until it fails and it exits condition 2 and print output,
# then again it will go to  condition 1, this process repeats until the condition 1 fails and print the output
```

```python
# end="" will stop printing the output in next line and print side by side
i=1 
while i<=5:
    print('Data Science',end=" ")
    j=1
    while j<=4:
        print('Technology',end=" ")
        j=j+1
    i=i+1
    print()
```

```python
i = 1          # while loop using numbers
while i<=2:
    j = 0
    while j<=2:
        print(i*j,end=" ")
        j+=1
    print()
    i+=1
```

```python
i=1            # while loop using numbers
while i<=4:
    j=0
    while j<=3:
        print(i*j, end=" ")
        j+=1
    print()
    i+=1 
```

```python
i=10           # while else loop using numbers
while i<=4:
    j=0
    while j<=3:
        print(i*j, end=" ")
        j+=1
    print()
    i+=1 
else:
    print('condition does not match')
```

# For Loop

```python
# while loop will work with iteration or condition 
# for loop will work with data sequence like list, string, integer
```

```python
name = 'SIVA'
for i in name:
    print(i)
```

```python
name = 'SIVA'
for i in name:
    print(i, end="")
```

```python
name1=[2,4.4,'SIVA']
for i in name1:
    print(i)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(10,50,5):
    print(i)
```

```python
# Print Numbers Divisible By 5
for i in range(1,51):
    if i%5==0:
        print(i)
```

```python
# Print Numbers not Divisible By 5
for i in range(1,51):
    if i%5!=0:
        print(i,end=" ")
```

```python
for i in range(1,20):   # Break the loop after reaching 8
    if i==8:
        break
    print(i)
```

```python
for i in range(1,11):    # continue will skip the number 8 and print the rest
    if i==8:
        continue
    print(i)
```
