```python
stats['myCalc']=stats['BirthRate']*stats['InternetUsers']   # now you can see mycalc column is added to your existing dataframe
stats.head()
```

# How to drop a column

```python
stats=stats.drop('myCalc',axis=1)    # axis= 0 means row
stats                                # axis=1 means column
```

```python
import matplotlib.pyplot as plt      # For Importing Libraries
import seaborn as sns

%matplotlib inline                   
plt.rcParams['figure.figsize']=6,2   # Size of the Graph  

import warnings
warnings.filterwarnings('ignore')    # will ignore os warnings
```

```python
stats["InternetUsers"]
```

```python
vis1=plt.distplot(stats["InternetUsers"])
plt.show()     # in distplot if we remove t we will get different graph 
               # distribution plot is advanced graph which you cant find in matplotlib
```

```python
vis1=sns.distplot(stats["InternetUsers"])  # distplot
plt.show()
```

```python
vis2=sns.displot(stats["InternetUsers"])    # displot
plt.show()
```

```python
vis3=sns.distplot(stats["InternetUsers"],bins=10)   # bins =10 means no of vertical column bars
plt.show()
```

```python
vis4=sns.boxplot(data=stats,x="IncomeGroup",y="BirthRate")    # Boxplot 
plt.show()   # The Graph Consists of 1) Outliers, 2) Maxline, 3) Whisker Plot, 4) Average Line, 5) Minimun Line
```

```python
vis5=sns.lmplot(data=stats,x="InternetUsers",y="BirthRate")  # lmplot = linear model plot
plt.show()              
```

```python
vis6=sns.lmplot(data=stats,x="InternetUsers",y="BirthRate",fit_reg=False)   # fit_reg = regression line
plt.show()                   # fit_reg = False Means Line Disappear
                             # Fit_reg = True means line appear
                             # by default fit_reg is true
```

```python
vis5=sns.lmplot(data=stats,x="InternetUsers",y="BirthRate",
                fit_reg=True, hue ='IncomeGroup')
plt.show()        
                   # This will give you graph different colours
```
