# Global Vs Local Variable

```python
a=10                   # Global Variable
def something():
    b=15               # Local Variable
    print('in function',b)

print('out function',a) 

#if you dont call local variable it will only print global variable
```

```python
a=10
def something():
    b=15
    print('in function',b)
something()
print('out function',a)
```

```python
# we can make a global variable into local but you cant make a local into global
 
a=10
def something():
    global a      # Here i have made global into local
    b=15
    print('in function',b)
    print('global variable',a)
something()
print('out function',a)
```

```python
x=10
def some():
    global x  # here we use 'global function' as local without a local variable
    x+=5      
some()
print(x)
```

```python
x=10             # here we are not defining any variable
def some():      # just access and modify the global variable using 'globals' function
    globals()['x']+=5 
some()
print(x)
```

# Fibonacci Sequence

```python
# sum of first two numbers is the next number
def fib(n):
    a=0
    b=1
    print(a)
    print(b)
    for i in range(0,n):
        c=a+b
        a=b
        b=c
        print(c)
fib(10)        
```

# Factorial Number

```python
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f
x=6
result=fact(x)
print(result)
```

# Filter

```python
# Find even numbers using function and list
def is_even(n):
    return n%2==0
nums=[5,4,6,3,2,8,10,7,1]
evens=list(filter(is_even,nums))
print(evens)
```

```python
# Find odd numbers using function and list
def is_odd(n):
    return n%2!=0
    nums=[5,4,6,3,2,8,10,7,1]
odds=list(filter(is_odd,nums))
print(odds)
```

```python
# Find even numbers using lambda function
nums=[5,4,6,3,2,8,10,7,1]
evens=list(filter(lambda n: n%2==0,nums))
print(evens)
```

```python
# Find odd numbers using lambda function
nums=[5,4,6,3,2,8,10,7,1]
odds=list(filter(lambda n: n%2!=0,nums))
print(odds)
```

```python
# find both even and odd numbers using lambda function
nums=[5,4,6,3,2,8,10,7,1]
evens=list(filter(lambda n: n%2==0,nums))
odds=list(filter(lambda n: n%2!=0,nums))
print(evens)
print(odds)
```

# Map

```python
# map using function
def update(n):
    return n+2
nums=[3,2,6,8,4,6,2,9]
evens=list(filter(is_even,nums))
double=list(map(update,evens))
print(evens)
print(double)
```

```python
# map using lambda function
nums=[3,2,6,8,4,6,2,9]
evens=list(filter(is_even,nums))
double=list(map(lambda n: n*2,evens))
print(evens)
print(double)
```

# Reduce

```python
# reduce with function
from functools import reduce
def add_all(a,b):
    return a+b
nums=[3,2,6,8,4,6,2,9]
evens=list(filter(is_even,nums))
double=list(map(lambda n: n*2,evens))
sums=reduce(add_all,double)
sums
print(sums)
```

```python
# reduce with lambda
from functools import reduce
nums=[3,2,6,8,4,6,2,9]
evens=list(filter(is_even,nums))
double=list(map(lambda n: n*2,evens))
sums=reduce(lambda a,b:a+b,double)
print(evens)
print(double)
print(sums)
```

# Decorator

```python
def div(a,b):
    print(a/b)
div(4,2)    
```

```python
def div(a,b):
    print(a/b)
div(2,4)    # even though i swap values i need same output that is 2.0 how?
```

```python
def div(a,b):
    if a<b:
        a,b=b,a
    print(a/b)
div(2,4)  # if i enter 2,4 or 4,2 answer is same. so this is decorator
```

```python
def my_decorator(func):
    def wrapper():
        print("something is happening before the function is called.")
        print("something is happening after the function is called.")
    return wrapper
@my_decorator
def say_hello():
    print("Hello")
say_hello()    
```

```python
def my_decorator(func):
    def wrapper():
        print("something is happening before the function is called.")
        print("something is happening after the function is called.")
        func()
    return wrapper
@my_decorator
def say_hello():
    print("Hello")
say_hello()    

# Output is same but new line added
```
