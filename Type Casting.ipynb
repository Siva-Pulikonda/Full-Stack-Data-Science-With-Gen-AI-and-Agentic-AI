# Type Casting 

# 1) Converting all data types to integers

```python
int(2.4)      # float to integer possible
```

```python
int(2.4,3.5)  # error because you cant declare two arguments while you do type casting 
```

```python
int(True)
```

```python
int(False)
```

```python
int(True,False)  # Two Arguments not possible
```

```python
int('10')   # If string is a number, type casting is valid
```

```python
int('ten')   # if the string is not a number then Type Casting is invalid
```

```python
# Number String to integer is valid but Text String to Integer is invalid
```

```python
int(1+2j)  # complex to integer is not possible
```

```python
# conclusion
# in type casting we cannot cast or convert string to integer, complex to integer, and cannot define two arguments 
```
