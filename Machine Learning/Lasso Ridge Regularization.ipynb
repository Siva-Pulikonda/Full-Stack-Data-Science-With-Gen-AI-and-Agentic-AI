```python
import pandas as pd
import numpy as np

# import graphical plotting libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# import linear regression machine learning libraries
from sklearn import preprocessing
# from sklearn.preprocessing import polynomialfeatures
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score
```

```python
dataset=pd.read_csv(r"C:\Users\sivap\OneDrive\Desktop\S\FSDS\December\Dec 18 2025\lasso, ridge, elastic net\TASK-22_LASSO,RIDGE\car-mpg.csv")
```

```python
dataset    # goal is to predict mpg (miles per gallon)
```

```python
# Drop car name
# replace origin into 1 2 3 .. dont forget to get dummies
# replace ? with nan
# replace all nan with median 

dataset=dataset.drop(['car_name'],axis=1)    # car_name is text → not useful for regression so we drop it 
dataset['origin']=dataset['origin'].replace({1: 'america', 2: 'europe',3: 'asia'})   # one hot encoder
dataset=pd.get_dummies(dataset,columns=['origin'],dtype=int)
dataset=dataset.replace('?',np.nan)                            # ?' → replaced with NaN, Missing values → filled with median. why median = Robust to outliers (better than mean)
dataset=dataset.apply(lambda x:x.fillna(x.median()), axis=0)
```

```python
dataset
```

```python
dataset=dataset.apply(pd.to_numeric,errors='ignore')

# fill missing values with median only for numeric columns
numeric_cols=dataset.select_dtypes(include=[np.number]).columns
dataset[numeric_cols]=dataset[numeric_cols].apply(lambda x: x.fillna(x.median()))
```

```python
dataset.head()
```

```python
x=dataset.drop(['mpg'],axis=1)   # independant variable
y=dataset[['mpg']]     # dependant variable
```

```python
# scaling the data
x_s=preprocessing.scale(x)
x_s=pd.DataFrame(x_s, columns=x.columns) # converting scaled data into dataframe

y_s=preprocessing.scale(y)
y_s=pd.DataFrame(y_s, columns=y.columns) # ideally train test data should be in columns

# preprocessing.scale is used for scaling all values between 0 to 1
```

```python
x_s
```

```python
y_s
```

```python
dataset.shape
```

```python
# split into train_test_data
x_train, x_test, y_train, y_test =train_test_split(x_s,y_s,test_size=0.2,random_state=0)
x_train.shape
```

# 2) simple linear model 

```python
#Fit simple linear model and find coefficients
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)

for idx, col_name in enumerate(x_train.columns):            # Shows impact of each feature on mpg
    print('The coefficient for {} is {}'.format(col_name, regression_model.coef_[0][idx]))

intercept = regression_model.intercept_[0]
print('The intercept is {}'.format(intercept))
```

# 2) b) Regularization Ridge Regression

```python
#alpha factor here is lambda (penalty term) which helps to reduce the magnitude of coeff

ridge_model = Ridge(alpha = 0.4)         # Ridge adds penalty: Reduces coefficient size, Prevents overfitting
ridge_model.fit(x_train, y_train)

print('Ridge model coef: {}'.format(ridge_model.coef_))
#As the data has 10 columns hence 10 coefficients appear here
# We Reduced the coeff values using ridge
```

# 2) c) Regularized lasso regression 

```python
#alpha factor here is lambda (penalty term) which helps to reduce the magnitude of coeff

lasso_model = Lasso(alpha = 0.1)          # Lasso penalty:Some coefficients → exactly 0, Performs feature selection
lasso_model.fit(x_train, y_train)

print('Lasso model coef: {}'.format(lasso_model.coef_))
#As the data has 10 columns hence 10 coefficients appear here
# here you can see most of high coeff has been scaled down to 0 and less numbers
```

```python
#Model score - r^2 or coeff of determinant
#r^2 = 1-(RSS/TSS) = Regression error/TSS    # R² Formula: 1 → perfect model, 0 → useless model, Negative → worse than mean

#Simple Linear Model
print(regression_model.score(x_train, y_train))
print(regression_model.score(x_test, y_test))

print('**********************')
#Ridge
print(ridge_model.score(x_train, y_train))
print(ridge_model.score(x_test, y_test))

print('**********************')
#Lasso
print(lasso_model.score(x_train, y_train))
print(lasso_model.score(x_test, y_test))
```
