# Data Cleaning Or Data Cleansing

```python
emp
```

```python
emp['Name']
```

```python
emp['Name']=emp['Name'].str.replace(r'\W','',regex=True) # It removes all non-alphanumeric characters from the Name column like (!, *, ?, >, ^, %) 
```

```python
emp['Name']
```

```python
emp['Domain']    # Domain also contains non alphabetic charecters 
```

```python
emp['Domain']=emp['Domain'].str.replace(r'\W','',regex=True)  
```

```python
emp['Domain']     # now you can see after using regex we cleaned the data
```

```python
emp['Location']
```

```python
emp['Location']=emp['Location'].str.replace(r'\W','',regex=True)
```

```python
emp['Location']
```

```python
emp['Age']
```

```python
emp['Age']=emp['Age'].str.replace(r'\W','',regex=True)    # we didnot get the required output using regex
# a string contains both number and text so to extract the digit we use this code                                   
```

```python
emp['Age']    
```

```python
emp['Age']=emp['Age'].str.extract('(\\d+)')  
```

```python
emp['Age']
```

```python
emp['Salary']
```

```python
emp['Salary']=emp['Salary'].str.replace(r'\W','',regex=True)  # Salary column is cleaned using regex
```

```python
emp['Salary']
```

```python
emp['Exp']
```

```python
emp['Exp']=emp['Exp'].str.extract('(\\d+)')  
```

```python
emp['Exp']
```

```python
# Now we are going to store this clean data in another copy 

clean_data=emp.copy()
```

```python
clean_data['Age']=clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
```

```python
clean_data['Exp']
```

```python
clean_data['Exp']=clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data['Exp']
```

```python
clean_data['Location']
```

```python
clean_data['Location']=clean_data['Location'].fillna(clean_data['Location'].mode()[0])
```
